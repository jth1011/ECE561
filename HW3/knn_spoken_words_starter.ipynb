```python
import numpy as np
import urllib
import pickle
import librosa;
import IPython
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Download custom spoken words dataset
x_train, x_test, y_train, y_test = pickle.load(urllib.request.urlopen("https://uwmadison.box.com/shared/static/6s7kzx5tnhn4b89kbecwhj5dx5nxmwge.pickle"))

print(np.shape(x_train))
```

```python
print(y_train[13])
IPython.display.Audio(x_train[13], rate = 16000)
```

```python
nperseg = 256
noverlap = 64

n_fft = 1024
hop_length = 256

X_train = []
for i in range(len(x_train)):
    S = librosa.feature.melspectrogram(x_train[i].astype(float), sr=32000, n_fft = n_fft, hop_length = hop_length)
    S_DB = librosa.power_to_db(S, ref=np.max)
    X_train.append(S_DB)

X_test = []
for i in range(len(x_test)):
    S = librosa.feature.melspectrogram(x_test[i].astype(float), sr=32000, n_fft = n_fft, hop_length = hop_length)
    S_DB = librosa.power_to_db(S, ref=np.max)
    X_test.append(S_DB)
    
print(np.shape(X_test))
print(np.shape(X_train))
```

```python
import librosa.display

for i in range(10):
    librosa.display.specshow(X_test[i], sr=32000, x_axis='time', y_axis='mel')
    plt.title(str(y_test[i]))
    plt.show()
    IPython.display.display(IPython.display.Audio(x_test[i],rate = 16000))
```
